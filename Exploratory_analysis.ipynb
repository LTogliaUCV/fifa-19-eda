```python
import pandas as pd
import matplotlib.pyplot as plt

# Carga el archivo CSV en un DataFrame de Pandas
df = pd.read_csv("kl.csv", encoding='latin-1')

# Elimina los caracteres no numricos de la columna "Wage"
df['Wage'] = df['Wage'].str.replace('K', '').str.replace('\x80', '')

# Convierte la columna "Wage" a un tipo numrico
df['Wage'] = pd.to_numeric(df['Wage'])

# Calcula la media y desviacin estndar del salario semanal
mean_wage = df['Wage'].mean()
std_wage = df['Wage'].std()

# Calcula los percentiles del salario semanal
percentiles_wage = df['Wage'].quantile([0.25, 0.5, 0.75])

# Identifica los jugadores cuyo salario semanal est a ms de 2 desviaciones estndar de la media
outliers_wage = df[(df['Wage'] > mean_wage + 2 * std_wage) | (df['Wage'] < mean_wage - 2 * std_wage)]

# Crea un grfico de barras para mostrar los percentiles del salario semanal
plt.bar(percentiles_wage.index, percentiles_wage.values)
plt.title("Percentiles del salario semanal")
plt.xticks(percentiles_wage.index, [f'{p*100}%' for p in percentiles_wage.index])
plt.xlabel("Percentiles")
plt.ylabel("Salario semanal")
plt.show()

# Crea un grfico de dispersin para mostrar los jugadores con salario atpico
plt.scatter(df['Wage'], df.index)
plt.scatter(outliers_wage['Wage'], outliers_wage.index, color='red')
plt.title("Salario semanal de los jugadores")
plt.xlabel("Salario semanal")
plt.ylabel("ndice del jugador")
plt.show()
```

```python
import pandas as pd
from prettytable import PrettyTable
import matplotlib.pyplot as plt

# Carga el archivo CSV en un DataFrame de Pandas
df = pd.read_csv("kl.csv", encoding='latin-1')
# Elimina los caracteres no numricos de la columna "Wage"
df['Wage'] = df['Wage'].str.replace('K', '').str.replace('\x80', '')

# Convierte la columna "Wage" a un tipo numrico
df['Wage'] = pd.to_numeric(df['Wage'])

# Convierte la columna "Wage" a un tipo numrico
df['Wage'] = pd.to_numeric(df['Wage'])

# Calcula la media y desviacin estndar del salario semanal
mean_wage = df['Wage'].mean()
std_wage = df['Wage'].std()

# Calcula los percentiles del salario semanal
percentiles_wage = df['Wage'].quantile([0.25, 0.5, 0.75])
percentiles_wage = percentiles_wage.round(2)

# Identifica los jugadores cuyo salario semanal est a ms de 2 desviaciones estndar de la media
outliers_wage = df[(df['Wage'] > mean_wage + 2 * std_wage) | (df['Wage'] < mean_wage - 2 * std_wage)]

# Crea una tabla para mostrar las estadsticas del salario semanal
table = PrettyTable()
table.field_names = ["Estadsticas del salario semanal", ""]
table.add_row(["Media", f"{mean_wage:.2f}"])
table.add_row(["Desviacin estndar", f"{std_wage:.2f}"])
table.add_row(["Percentiles", f"{percentiles_wage}"])

# Imprime la tabla
print(table)

# Crea un grfico de barras para mostrar los percentiles del salario semanal
plt.bar(percentiles_wage.index, percentiles_wage.values)
plt.title("Percentiles del salario semanal")
plt.xticks(percentiles_wage.index, [f'{p*100}%' for p in percentiles_wage.index])
plt.xlabel("Percentiles")
plt.ylabel("Salario semanal")
plt.show()
```
